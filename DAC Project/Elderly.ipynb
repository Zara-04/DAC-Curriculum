```python
# Reload the newly uploaded dataset
file_path = 'cleaned_file.csv'
data = pd.read_csv(file_path)

# Filter out data for people aged 65 and above and label them as 'Elderly'
data['Age Group'] = data['Age'].apply(lambda x: 'Elderly' if x >= 65 else 'Non-Elderly')

# Filter only elderly data
elderly_data = data[data['Age Group'] == 'Elderly']

# Save or display the elderly dataset
elderly_data.to_csv('elderly_data.csv', index=False)
print("Filtered dataset saved as 'elderly_data.csv'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Reloading the dataset
file_path = 'cleaned_file.csv'  # Update the path if necessary
data = pd.read_csv(file_path)

# Filter out elderly data (age >= 65)
elderly_data = data[data['Age'] >= 65]

# Perform exploratory data analysis (EDA)
# Distribution of ages among the elderly
plt.figure(figsize=(8, 5))
sns.histplot(elderly_data['Age'], kde=True, bins=10, color='skyblue')
plt.title("Age Distribution of Elderly Patients")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Gender distribution among the elderly
plt.figure(figsize=(8, 5))
sns.countplot(data=elderly_data, x='Gender', palette='pastel')
plt.title("Gender Distribution of Elderly Patients")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# Most common medical conditions among the elderly
plt.figure(figsize=(10, 6))
elderly_data['Medical Condition'].value_counts().plot(kind='bar', color='lightcoral')
plt.title("Common Medical Conditions Among Elderly Patients")
plt.xlabel("Medical Condition")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
